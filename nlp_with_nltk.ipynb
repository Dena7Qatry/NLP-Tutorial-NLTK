### **WordNetLemmatizer in NLTK**

The `WordNetLemmatizer` is a class within the **NLTK** library in Python. Its primary function is to perform **lemmatization**, which is the process of **reducing words to their base or dictionary form**, known as the ***lemma***.
 - Unlike stemming, which often just **chops off suffixes** and can result in non-words, **lemmatization** uses a **vocabulary** and **morphological** analysis of words to return a valid word from the dictionary.

**For example:**

- Stemming might reduce `"studying"`, `"studies"`, and `"studied"` to the stem **`"studi"`**.
- Lemmatization would reduce `"studying"`, `"studies"`, and `"studied"` to the lemma **`"study"`**.

---

#### **How does `WordNetLemmatizer` work?**

The `WordNetLemmatizer` utilizes the **WordNet** lexical database. WordNet is a large English lexical database that groups English words into sets of synonyms called ***synsets***, and provides:

- Short definitions
- Usage examples
- Various semantic relations between synsets

When you use `lemmatize()` with the `WordNetLemmatizer`, it looks up the word in WordNet and attempts to find its base form based on the specified **part of speech**.

---

#### **Importance of Part of Speech (POS) Tagging**

A key aspect of using `WordNetLemmatizer` effectively is providing the correct **part of speech (POS)** tag for the word you are trying to lemmatize. The lemma of a word can change depending on its POS:

- The lemma of `"better"` as an **adjective** is `"good"`.
- The lemma of `"better"` as a **verb** is `"better"`.

The `lemmatize()` method accepts a `pos` argument, where you can specify the POS using these codes:

- `'n'` for **noun**
- `'v'` for **verb**
- `'a'` for **adjective**
- `'r'` for **adverb**

>  If no part of speech is provided, the `WordNetLemmatizer` defaults to assuming the word is a **noun** (`'n'`). Therefore, performing POS tagging prior to lemmatization often yields more accurate results.

---

#### Summary

The `WordNetLemmatizer` is a powerful tool in NLTK that enables **accurate lemmatization** by:

- Leveraging the **WordNet** database
- Allowing for **part-of-speech specification**


```
from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()
word_list = ["study","studying","studies","studied"]

for w in word_list:
    print(lemma.lemmatize(w ,pos="v"))
```

### **Chinking in NLTK**

This code snippet demonstrates a concept in **NLP** called **chinking**, using the **NLTK** library.

- **Chinking** is the process of **excluding a pattern of tags** from a previously defined chunk.
- It allows for **more precise control over chunked phrases** by removing specific elements based on their part-of-speech tags.


```
#Chinking example :
# * - 0 or more repetations
# + - 1 or more repetations

#Here we are taking the whole string and then
#excluding adjectives from that chunk.

grammar = r""" NP: {<.*>+}
               }<JJ>+{"""

#  A grammar is defined using a multi-line string (indicated by triple quotes """).
# The r before the opening triple quote means it's a raw string, which is often useful for regular expressions to avoid issues with backslashes.


# Creating parser :
parser = nltk.RegexpParser(grammar)

#parsing string :
output = parser.parse(tagged_words)
print(output)

#To visualize :
#output.draw()
```

### ***Explanation of the Chinking Code in NLTK***

The code performs the following steps:

---

#### 1. Defines a Grammar for Chinking:

- A grammar is defined using a regular expression pattern stored in the `grammar` variable.
- The line `NP: {<.*>+}` defines an initial chunking rule:
  - It creates a chunk labeled `NP` (typically meaning *Noun Phrase*), capturing **one or more** (`+`) of **any tag** (`<.*>`).
  - This effectively creates one large chunk containing **all words and their POS tags**.

- The line `}<JJ>+{` defines the **chinking rule**:
  - The curly braces `}` and `{` denote the start and end of the **exclusion pattern**.
  - `<JJ>+` matches **one or more adjectives**.
  - This rule removes any adjective(s) from the NP chunk.

---

#### 2. Creates a Regular Expression Parser:

- `parser = nltk.RegexpParser(grammar)` creates a **parser** using the defined grammar.
- The parser applies **chunking and chinking rules** to POS-tagged text.

---

#### 3. Parses the Tagged Words:

- `output = parser.parse(tagged_words)` runs the parser on a list of `tagged_words`.
- These are words with their POS tags (from a previous tagging step).
- The parser returns a **parse tree**, applying the chunking and chinking logic.

---

#### 4. Prints the Output:

- `print(output)` displays the **chunked structure**.
- It shows the words, their tags, and the **effect of chinking** (adjectives removed from NP chunks).

---

#### 5. Visualizes the Output (Commented Out):

- `#output.draw()` is a **commented-out line**.
- If uncommented and executed in a GUI-enabled environment (with NLTK visualization tools installed), it displays a **graphical parse tree** to visualize the chunks and chinks.

### **Named Entity Recognition (NER)**

**Named Entity Recognition (NER)** is a task in **NLP** that aims to **locate and classify named entities** in text into predefined categories such as:

- Names of **persons**
- **Organizations**
- **Locations**
- Expressions of **time**
- **Quantities**
- **Monetary values**
- **Percentages**, etc.


```
#Sentence for NER :
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#Tokenizing words :
tokenized_words = word_tokenize(sentence) # word_tokenize function from nltk is used to break down the sentence into individual words or tokens

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words)

# Download the required named entity chunker resource
import nltk
nltk.download('maxent_ne_chunker_tab') # downloads the maxent_ne_chunker_tab resource, which is necessary for the named entity chunking process.

#Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=False) # nltk.ne_chunk() is used to perform named entity chunking

# The argument binary=False tells the function to classify the named entities into different categories (like PERSON, GPE, ORGANIZATION) rather than just labeling them as a generic 'NE' (Named Entity).
print(N_E_R)

```

### **Accessing Synsets with WordNet**.

**Synsets** (short for *synonym sets*) are groups of **cognitive synonyms** that represent a specific concept and are used for **lexical semantic analysis** in **WordNet**.


**WordNet** is a large **lexical database** of English **nouns**, **verbs**, **adjectives**, and **adverbs**.

It groups words into sets of **cognitive synonyms** called **synsets**, each representing a distinct concept. These synsets are interlinked by means of **conceptual-semantic and lexical relations**, making **WordNet** a powerful tool for **semantic analysis** in NLP tasks.



```
#Import wordnet :
from nltk.corpus import wordnet

for words in wordnet.synsets("Fun"): # This function takes a word as input (in this case, "Fun") and returns a list of all the synsets that contain that word
    print(words)
```

## **Function above**
### **Iterating Through Synsets in WordNet**

This code iterates through all the **synsets** found in WordNet for the word `"Fun"`.

The function `wordnet.synsets("Fun")` returns a **list of synset objects**, each representing a different **meaning or sense** of the word.

Inside the loop, for each synset:

- `print(words)` displays the **representation of that synset object**.
- This typically includes the **word**, its **part of speech**, and a **unique identifier**.

This process helps explore the various semantic interpretations of a word in natural language.




---



```
#Word meaning with definitions :
for words in wordnet.synsets("Fun"):
    print(words.name())
    print(words.definition())
    print(words.examples())

    for lemma in words.lemmas():
        print(lemma)
    print("\n")
```

### **Finding Synonyms with WordNet**

WordNet includes lexical relationships such as **synonyms**, **antonyms**, **hypernyms**, and **hyponyms**, making it useful for various semantic analysis tasks in NLP.


By accessing synsets and their associated **lemmas**, you can retrieve alternative words that share similar meaningsuseful for tasks like **query expansion**, **text normalization**, and **semantic similarity analysis** in NLP.


```
#Finding synonyms :

#Empty list to store synonyms :
synonyms = []

for words in wordnet.synsets('Fun'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

### **Finding Antonyms with WordNet**





```
#Finding antonyms :

#Empty list to store antonyms :
antonyms = []

for words in wordnet.synsets('Natural'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

#Print antonyms :
antonyms
```

```
#Finding synonyms and antonyms :

#Empty lists to store synonyms/antonynms :
synonyms = []
antonyms = []

for words in wordnet.synsets('New'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

#Print lists :
print(synonyms)
print("\n")
print(antonyms)
```

### **Bag of Words (BoW) Representation**

The Bag of Words model is a fundamental technique in NLP that simplifies text data by:

+ Treating each document (like a sentence) as an unordered collection of words.
+ Focusing on the frequency of each word within the document.
+ Ignoring grammar and word order.


==> This representation is valuable for various NLP tasks, such as classifying text or finding similar documents.




```
# Import required libraries :
from sklearn.feature_extraction.text import CountVectorizer

# Text for analysis :
sentences = ["Jim and Pam travelled by the bus:",
             "The train was late",
             "The flight was full.Travelling by flight is expensive"]

# Create an object :
cv = CountVectorizer() # This creates a list of strings called sentences. This list contains the text data that we want to convert into a Bag of Words representation.

# Generating output for Bag of Words :
B_O_W = cv.fit_transform(sentences).toarray()

# Total words with their index in model :
print(cv.vocabulary_)
print("\n")

# Features :
# Changed get_feature_names() to get_feature_names_out() for compatibility with newer scikit-learn versions
print(cv.get_feature_names_out())
print("\n")

# Show the output :
print(B_O_W)
```

### **Text Representation with TF-IDF**

**TF-IDF** is a numerical statistic that reflects how important a word is to a document in a collection or corpus. It is commonly used as a **weighting factor** in tasks such as:

- **Information retrieval**
- **Text mining**
- **Text classification**

TF-IDF helps reduce the weight of commonly used words and highlight terms that are more **informative** for a particular document.


```
#Import required libraries :
from sklearn.feature_extraction.text import TfidfVectorizer

#Sentences for analysis :
sentences = ['This is the first document','This document is the second document']

#Create an object :
vectorizer = TfidfVectorizer(norm = None)

#Generating output for TF_IDF :
X = vectorizer.fit_transform(sentences).toarray()

#Total words with their index in model :
print(vectorizer.vocabulary_)
print("\n")

#Features :
# Changed get_feature_names() to get_feature_names_out() for compatibility with newer scikit-learn versions
print(vectorizer.get_feature_names_out())
print("\n")

#Show the output :
print(X)
```
